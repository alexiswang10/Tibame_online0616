```python
# 查詢一段
scores = [30, 50, 80, 60, 90]
print(scores[1:3])
print(scores[1:-1])
# 前(末)取段
print(scores[:3])
print(scores[1:])

a="_"*10
print(a)
```

```python
# 跳號
print(scores[0:4:2])
# 取奇數 取偶數
print(scores[::2], scores[1::2])

# 反向跳號 (從右跳到左)
print(scores[4:0:-2])
print(scores[-1:0:-2])
print(scores[::-1])
```
