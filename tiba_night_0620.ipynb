```
python群集型態:
1. list(清單): 同類型的集合
   [20, 30, 40]
   查詢: list[key]
   * 有序(key=0,1,2...)
2. dict(字典): 聚集小東西變成一個複雜的東西
   {"name":"Elwing", "height":175}
   查詢: dict[key]
   * 無序(key自訂)
!!! JSON格式: list+dict
3. set(集合): 很類似list, 但是裡面的東西絕不重複的集合
   {"周", "王", "黃"}
   * 無序(沒有供使用者取用的key)
4. tuple: 簡化版本的dict
   e.g. 假設我今天需要來表示顏色 {"R":255, "G":128, "B":120} -> (255, 128, 120)
   * 有序(key=0,1,2...)
   !!! 唯一可以當成字典的key或者可以放到set裡(hashable: 不可變動的)
```

````python
s1 = {"周", "王", "黃", "周"}
s2 = {"王", "何"}
# 交集(&)(and): 我們都有的
# 聯集(|)(or): 所有的
# 減集(-): 我-你有的
print(s1 & s2)
print(s1 | s2)
print(s1 - s2)
print(s2 - s1)

print(len(s1))

# add/discard
s1.add("李")
print(s1)

# remove和discard的差別:
# 如果東西不在
# remove會報錯誤
# discard不做事
target = "劉"
# s1.remove(target)
s1.discard(target)
print(s1)

print("周" in s1)
````

````python
# 發現循環也要停止
a, b = 44, 7
result = str(a // b) + "."
a = a % b

# 如果這個商已經看過, 後面必定重複, 就停下
seen = set()
i = 0
while i < 10:
    # 2(a) -> 20
    a10 = a * 10
    if a10 in seen:
        print("循環")
        break
    else:
        seen.add(a10)
    # 20 // 7
    result = result + str(a10 // b)
    print(result)
    # 20 % 7
    a = a10 % b
    if a == 0:
        print("整除")
        break
    i = i + 1
````

````python
import random
def generate_lottery():
    # 買一張彩券(49 選 6)
    lottery = set()
    # 你要寫個迴圈, 一職add近去, 直到裡面有6個為止
    while len(lottery) < 6:
        n = random.randint(1, 49)
        lottery.add(n)
    return lottery

# 產生頭彩的彩券
prize = generate_lottery()
print(prize)
# 產生特別號(產生一個不在prize裡的號碼當特別號)
while True:
    special = random.randint(1, 49)
    if not special in prize:
        break
print("特別號是:", special)

# 瘋狂做
ticket = generate_lottery()
print(ticket)
````
