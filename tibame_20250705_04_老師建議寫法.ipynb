````python
def flow(n):
    return pd.Series({"a":n, "b":2*n})
df["rating"].apply(flow)
````

````python
def flow(data):
    data_split = data.split(" / ")
    if len(data_split) == 1:
        area = data_split[0]
        genre = "-"
    else:
        area = data_split[0]
        genre = data_split[1]
    return pd.Series({"area":area, "genre":genre})

df[["area", "genre"]] = df["area_genre"].apply(flow)
df
````

````python
holiday_keys = [
    "星期一",
    "星期二",
    "星期三",
    "星期四",
    "星期五",
    "星期六",
    "星期日",
]

def flow(data):
    holiday = {}
    for key in holiday_keys:
        holiday[key] = "Y"

    for key in holiday:
        if key in data:
            holiday[key] = "N"
    return pd.Series(holiday)

df[holiday_keys] = df["holiday"].apply(flow)
df
````
