** 老師開示 **

````python
class MyList:
    def __init__(self):
        self.data = []

    def __round__(self):
        new = [round(d) for d in self.data]
        return new

l = MyList()
l.data.append(5.6)
l.data.append(200.213)
new = round(l)
new
````
