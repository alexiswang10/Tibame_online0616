```python
# 引進json套件做型態轉換
import json
import urllib.request as req

url = "https://www.google.com/doodles/json/2023/7?hl=zh-TW"

f = req.urlopen(url) # 從遠處打開檔案
s = f.read() # 讀取遠處檔案
pics = json.loads(s)
# pics = json.load(f)****1
# pics = json.loads(s)***2
# 1或2挑一個寫作型態傳換
# print(type(pics))

for p in pics:
    # print(type(p))
    # 下面圖片二中"字典型態"你的title和網址high_res_url
    print(p["title"])
    img = "https:" + p["high_res_url"]
    # 下面圖三中的
    print(img)
    print("-" * 30)
```
